# 1. Примеры создания массивов

## 1.1 Из обычных списков

```python
list_array = np.array([1, 2, 3, 4, 5])
print("Массив из списка:", list_array)
```

## 1.2 С использованием функций zeros, ones, arange

```python
zeros_array = np.zeros(5)  # массив из 5 нулей
ones_array = np.ones(20)    # массив из 20 единиц
arange_array = np.arange(0, 10, 2)  # массив от 0 до 10 с шагом 2

print("Zeros:", zeros_array)
print("Ones:", ones_array)
print("Arange:", arange_array)
```

# 2. Демонстрация атрибутов: форма и тип данных

```python
print("Форма массива list_array:", list_array.shape)
print("Тип данных массива list_array:", list_array.dtype)
```

# 3. Примеры индексирования и срезов

## 3.1 Получение отдельного элемента

```python
print("Первый элемент list_array:", list_array[0], 
      "Второй элемент list_array:", list_array[1])
```

## 3.2 Извлечение подмассива с помощью среза

```python
sub_array = list_array[1:4]  # элементы с индексами 1, 2, 3
print("Срез массива [1:4]:", sub_array)
```

# 4. Примеры векторных операций и базовых статистик

## 4.1 Поэлементное сложение массивов

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("Поэлементное сложение:", a + b)
```

## 4.2 Вычисление средней величины

```python
mean_value = np.mean(a)
print("Среднее значение массива a:", mean_value)
```
